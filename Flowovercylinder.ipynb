## Flow past a cylinder with circulation

```python
import sympy
from sympy.abc import x, y
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cylinder_stream_function(U=1, R=1,mu=1,gamma=1):
    r = sympy.sqrt(x**2 + y**2)
    theta = sympy.atan2(y, x)
    return U*r*sympy.sin(theta)-((mu*sympy.sin(theta))/(2*np.pi*r))-((gamma*sympy.log(r))/(2*np.pi))

def velocity_field(psi):
    u = sympy.lambdify((x, y), psi.diff(y), 'numpy')
    v = sympy.lambdify((x, y), -psi.diff(x), 'numpy')
    return u, v

def plot_streamlines(ax, u, v, xlim=(-1, 1), ylim=(-1, 1)):
    x0, x1 = xlim
    y0, y1 = ylim
    Y, X =  np.ogrid[y0:y1:100j, x0:x1:100j]
    ax.streamplot(X, Y, u(X, Y), v(X, Y), color='cornflowerblue')
    
        
psi = cylinder_stream_function()
u, v = velocity_field(psi)

xlim = ylim = (-1,1)
fig, ax = plt.subplots(figsize=(6, 6))
plot_streamlines(ax, u, v, xlim, ylim)

c = plt.Circle((0, 0), radius=0.4,fill=False )
ax.add_patch(c)

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.show()
#plt.savefig('fpac_1.png', format='png')
plt.savefig('fpac_1.eps', format='eps')
```

```python
xlim = ylim = (-1.5,1.5)
fig, ax = plt.subplots(figsize=(6, 6))
plot_streamlines(ax, u, v, xlim, ylim)
c = plt.Circle((0, 0), radius=0.4,fill=False )
ax.add_patch(c)
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.show()
plt.savefig('fpac_2', format='png')
```
